```python
# Impossible Event
print("Probability of getting 7 on a six-sided dice: 0")
print("\n-Impossible event")
```

```python
# Unlikely Event (Less likely)
import random

roll = random.randint(1, 6)
print("Probability of getting 1 on a six-sided dice:", 1 / 6)
print(f"Rolled {roll}")

if roll == 1:
    print("You rolled a one. (Unlikely)")
else:
    print("Try again.")
```

```python
# Even Chance Event
print("Probability of getting a heads or tails in a coin flip:", 1 / 2)

flip = random.choice(['Heads', 'Tails'])
print("The coin has landed on", flip)
print("\nIt's a likely event.")
```

```python
# Likely Event
print("Probability of getting any other number than a 6 on a dice:", 5 / 6)

roll = random.randint(1, 6)
if roll != 6:
    print("More likely.")
else:
    print("You got a 6. (Less likely event)")
```

```python
# Certain Event
print("Will the sun rise tomorrow?")
print("Probability : 1")
print("\nThis is a Certain Event.")
```
