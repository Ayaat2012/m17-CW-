```python
# Simulate Central Limit Theorem by generating samples of size 1, 10, 50 and 100 respectively 
# and plotting histograms for the same.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nums = [1, 10, 50, 100]
means = []

for i in nums:
    np.random.seed(1)
    x = [np.mean(np.random.randint(-40, 40, i)) for j in range(1000)]
    means.append(x)


# Plotting all the means in one figure
k = 0
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
for i in range(2):
    for j in range(2):
        ax[i, j].hist(means[k], 10, density=True)
        ax[i, j].set_title(label=nums[k])
        k += 1  #k = k+1

plt.show()
```
